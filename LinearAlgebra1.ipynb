```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t= np.linspace(0,40,100)
```

```python
dr = 2.5*t
```

```python
ds = 3 * (t-5)
```

```python
fig, ax = plt.subplots()
plt.title("A Bank Rober Catch")
plt.xlabel("time")
plt.ylabel("distance")
ax.set_xlim([0,40])
ax.set_ylim([0,100])
ax.plot(t, dr, c='green')
ax.plot(t, ds, c='red')
plt.axvline(x=30,color="purple", linestyle='--')
_ =plt.axhline(y=75,color="purple", linestyle='--')
```
